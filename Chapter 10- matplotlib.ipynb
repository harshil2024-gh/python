__[Task]__:The cost categories applied to a 9 dollar micro controller Engineering is 1.35 Manufacturing is 3.6 sales is 2.25 and profit is 1.8 create a pie chart that will show the cost breakdown as different size pieces also there will be an exploded view for profit.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,7))
cost_cat=np.array([1.35,3.6,2.25,1.8])
val=np.array(["Purchase","Manufacturing","Sales","Profit"])
exp=[0,0,0,0.1]
plt.pie(cost_cat,labels=val,explode=exp)
plt.legend(title="Cost Revenue")
plt.autopct("")
plt.tight_layout()
plt.show()
```
