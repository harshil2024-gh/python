# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,color='red',label="ABC") #We can also give hexcode instead of colour.
x1=np.array(["P","Q","R","S"])
y1=np.array([3,8,1,10])
plt.bar(x1,y1,color='#4CAF50',label="PQR")
plt.legend()
plt.show()
#For horizontal bar
# plt.barh(x,y)
# plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C"])
x1=np.array(["P","Q","R","S"])
y1=np.array([3,8,1,10])
plt.bar(x1,y1,color='#4CAF50',width=0.3) #We can control width of the bar by width attribute.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C"])
x1=np.array(["P","Q","R","S"])
y1=np.array([3,8,1,10])
plt.barh(x1,y1,color='purple',height=0.5) #We can control height of the horizontal bar by height attribute.
plt.show()
```

# PIE CHART

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
plt.pie(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
plt.pie(x,startangle=90) #startangle=90 will create chart from 90deg in anti-clockwise direction
plt.show()
```

```python
#Labels
import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apple","Banana","Cherry","Dates"]
plt.pie(x,labels=mylabels)
plt.show()
```

```python
#Explode
import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apple","Banana","Cherry","Dates"]
myexplode=[0.1,0,0,0]
plt.pie(x,labels=mylabels,explode=myexplode)
plt.show()
```

```python
#Shadow
import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apple","Banana","Cherry","Dates"]
myexplode=[0.1,0,0,0]
plt.pie(x,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

```python
#colours
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(9,9))
x=np.array([35,29,20,15])
mylabels=["Apple","Banana","Cherry","Dates"]
myexplode=[0.1,0,0,0]
color=["red","yellow","magenta","brown"]
plt.title("Fruit Production")
plt.pie(x,labels=mylabels,explode=myexplode,colors=color,autopct="%1.1f%%")
plt.legend(title="Fruits")
plt.show()
```
